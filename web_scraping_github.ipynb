```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ebay.de/sch/i.html?_from=R40&_trksid=p4432023.m570.l1313&_nkw=scooter&_sacat=0'

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
   
    # Find all the product titles
product_titles = soup.find_all('div',class_='s-item__title')
title =[]
for div in product_titles:
    # Extract the data or perform operations on each <div> element
    title.append(div.span.text.strip()) 
    

product_prices = soup.find_all('span', class_='s-item__price')
price =[]
for i in product_prices:
    price.append(i.text.strip()) 
    
    
product_feature = soup.find_all('span', class_='s-item__dynamic s-item__coupon')
feature = []
for i1 in product_feature:
    feature.append(i1.text.strip())    
    
seller_info =[]    
product_seller = soup.find_all('span', class_='s-item__seller-info-text')
for i2 in product_seller:
    seller_info.append(i2.text.strip())         
    
sub_Title_info =[]    
sub_Title_information = soup.find_all('div', class_='s-item__subtitle')
for i3 in sub_Title_information:
    sub_Title_info.append(i3.text.strip())
```

```python
data_scooter = list(zip(title, price,seller_info,sub_Title_info))
```

```python
# Define column names for the DataFrame
columns = ['Title', 'Price', 'Seller','sub_title']

# Create the DataFrame
df = pd.DataFrame(data_scooter, columns=columns)

# Print the DataFrame
df
```

```python
df[['Username', 'Feedback Score', 'Positive Feedback %']] = df['Seller'].str.split(r' \(|\) ', expand=True)     
```

```python
df.drop("Seller",axis=1,inplace = True)
```
